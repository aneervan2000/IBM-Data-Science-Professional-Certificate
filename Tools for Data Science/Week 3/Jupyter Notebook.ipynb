### The following code find the sum of n natural numbers

```python
n = int(input('Enter a number: '))
s = 0
for i in range(n + 1):
    s = s + i
print(s)
```

* **Formula:** (n*(n+1))/2
* (10 * 11)/2 = 55

>The above is the formula to find out the sum on any n natural numbers
